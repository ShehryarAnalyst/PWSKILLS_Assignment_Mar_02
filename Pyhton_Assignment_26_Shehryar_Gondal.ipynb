__Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.__

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is x_label")
plt.ylabel("This is y_ label")
plt.title("Title")
plt.show()
```

__Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])__

__Ans)__ The subplot() function is used to create multiple plots in a single figure. It is commonly used when we want to compare or analyze multiple datasets in a single visualization.

```python
#plot 1:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.figure(figsize=(10,8))
plt.subplot(4, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.figure(figsize=(10,8))
plt.subplot(4, 1, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.figure(figsize=(10,8))
plt.subplot(4, 1, 3)
plt.plot(x,y)

#plot 4:
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10,8))
plt.subplot(4, 1, 4)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Set titles for subplots
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Plot the lines
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add space between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# Show the plot
plt.show()
```

Each bar represents a category, and the height or length of the bar represents the value of the data for that category. Bar plots are used to visualize the distribution or comparison of categorical data.

Bar plots are useful when we want to compare the frequency, count, or proportion of different categories. They are commonly used in various fields such as business, finance, social sciences, and many more. For example, a bar plot can be used to compare the sales of different products or to compare the performance of different teams or individuals.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

__Ans)__ A box plot is a type of plot used to visualize the distribution of data and to identify outliers. It displays the five-number summary of the data (minimum, first quartile, median, third quartile, and maximum) as a box, with whiskers extending from the box to indicate the range of the data, and points outside the whiskers to represent outliers.

Box plots are used to compare the distribution of data between different groups or categories. They are particularly useful when we have a large amount of data and want to quickly identify the central tendency, spread, and potential outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Create box plot
fig, ax = plt.subplots()
ax.boxplot(data)

# Set labels
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
